1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples

The difference between **descriptive statistics** and **inferential statistics** lies in **what they do with data**:

---

**Descriptive Statistics**

Descriptive statistics **summarize and describe** the features of a dataset.
It deals with the **data you already have** — no assumptions beyond the dataset.

 Purpose:

To provide a clear picture of the collected data.

 Examples:

| Concept            | Example                                                    |
| ------------------ | ---------------------------------------------------------- |
| Mean (average)     | “The average score of students in the class is 76.”        |
| Median             | “The median salary in the company is ₹45,000 per month.”   |
| Standard deviation | “There is low variation in customer satisfaction ratings.” |
| Charts/Graphs      | Bar charts, pie charts, histograms, etc.                   |

 **Example in real life:**
A company surveys 200 customers and finds:

* Average age = 32 years
* 65% are female
* 80% are satisfied with the product

 These results **describe** the sample; they do not go beyond it.

---

 **Inferential Statistics**

Inferential statistics **make predictions or conclusions** about a larger population based on a **sample** of data.

 Purpose:

To **generalize** from a sample to a population.

 Examples:

| Concept              | Example                                                                             |
| -------------------- | ----------------------------------------------------------------------------------- |
| Hypothesis testing   | “Is the new medicine more effective than the old one?”                              |
| Confidence intervals | “We are 95% confident that the average monthly salary in India is ₹30,000–₹35,000.” |
| Regression           | Predicting sales based on advertising budget.                                       |
| Probability          | Forecasting election results based on sample voting data.                           |

 **Example in real life:**
A political analyst surveys **1,000 voters** and predicts:

* Candidate X will win the election with **52% of the total votes** nationwide.

 The conclusion is about **the whole population**, even though data was collected only from a **sample**.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling

**What is sampling in statistics?**

Sampling is the process of **selecting a subset of individuals from a larger population** in order to collect data and make conclusions about the **whole population** without studying every member.

---

 **Random Sampling**

Random sampling means **every member of the population has an equal chance of being selected**.

#### Example:

Selecting 50 students randomly from a school where every student has an equal probability of selection.

 Simple
 Reduces bias

---

 **Stratified Sampling**

Stratified sampling means the population is **divided into subgroups (strata)** based on characteristics like gender, age, department, etc., and **samples are taken from each subgroup proportionally**.

#### Example:

If a school has:

* 60% boys
* 40% girls

Then from a sample of 50 students:

* 30 boys
* 20 girls

 Ensures representation of all groups
 More accurate than random sampling when population is diverse

3. Define mean, median, and mode. Explain why these measures of central
tendency are importan

**Mean, Median, and Mode (Measures of Central Tendency)**

 **Mean (Average):**
The mean is the sum of all values divided by the number of values.
Example: (5 + 7 + 8 + 10) / 4 = **7.5**

**Median (Middle Value):**
The median is the middle value when data is arranged in ascending or descending order.
Example: In 3, 6, 8, 12, 15 → **8** is the median.

 **Mode (Most Frequent Value):**
The mode is the value that occurs **most often** in the dataset.
Example: In 2, 4, 4, 7, 9 → **4** is the mode.

---

**Why these measures are important**

They help summarize and understand the **central or typical value** of a dataset:

* They provide a **single representative value** for the data.
* Allow easy **comparison between datasets**.
* Useful in **decision-making** in fields like business, economics, education, and research.


4. Explain skewness and kurtosis. What does a positive skew imply about the
data

 **Skewness**

Skewness measures the **asymmetry** of a data distribution.

* **Symmetrical distribution** → data is evenly spread around the mean.
* **Skewed distribution** → data is stretched more to one side.

 **Kurtosis**

Kurtosis measures the **tailedness** or **peakedness** of a distribution.

* High kurtosis → sharp peak and heavy tails (more extreme values).
* Low kurtosis → flatter peak and light tails.

---

 **What does a positive skew imply?**

A **positive skew (right skew)** means:

* The **tail of the distribution extends to the right**.
* **Most values are concentrated on the left**, with a few large values to the right.
* Order of central tendency usually:
  **Mean > Median > Mode**

---


```python
#  Implement a Python program to compute the mean, median, and mode of a given list of numbers.

from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = mean(numbers)
median_value = median(numbers)
mode_value = mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
# Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Covariance
covariance = np.cov(x, y, ddof=0)[0][1]   # population covariance

# Correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
# Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Given Data")
plt.ylabel("Values")
plt.show()

# Detect outliers using IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
# You are working as a data analyst in an e-commerce company. Themarketing team wants to know if there is a relationship between advertising spend and daily sales

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

print("Correlation Coefficient:", correlation)
```
